## **Image Processing**

```python
# IMAGE RESIZE

img = cv2.imread("bird.jpg", 1)

half = cv2.resize(img, (0, 0), fx = 0.5, fy = 0.5)
bigger = cv2.resize(img, (1050, 1610))
 
 
Titles =["Original", "Half", "Bigger"]
images =[img, half, bigger]
count = 3
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```

```python
# IMAGE BLUR

# Average Blur 
avg_blur = cv2.blur(img,(7,7)) 
cv2.imshow('Average Blurring', avg_blur)

# Median Blur 
median = cv2.medianBlur(img, 25) 
cv2.imshow('Median Blurring', median)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# IMAGE ROTATION

(rows, cols) = img.shape[:2]

M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 90, 1) 
res = cv2.warpAffine(img, M, (cols, rows))

cv2.imshow('Rotate 90 degrees', res)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# EDGE DETECTION

img = cv2.imread("bird.jpg",0)

# Canny
canny = cv2.Canny(img,30,40)

# sobel-X
sobelX = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)

# cv2.imshow('sobelX', sobelX)

# cv2.waitKey(0)
# cv2.destroyAllWindows()

# sobel-Y
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)

plt.subplot(221),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(222),plt.imshow(canny,cmap = 'gray')
plt.title('Edge Image (Canny)'), plt.xticks([]), plt.yticks([])

plt.subplot(223),plt.imshow(sobelX, cmap='gray')
plt.title('Edge Image (Sobel-X)'), plt.xticks([]), plt.yticks([])

plt.subplot(224),plt.imshow(sobelY,cmap = 'gray')
plt.title('Edge Image (Sobel-Y)'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# CONVERT IMAGE TO OTHER COLOR SPACE

img = cv2.imread('bird.jpg',1)

convert_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
convert_rgba = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2.imshow('Convert to HSV', convert_hsv)
cv2.imshow('Convert to RGBA', convert_rgba)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Video Processing**

Biasanya OpenCV sering digunakan untuk mengambil frames dari sebuah video. Video tersebut dapat diambil dari sebuah file atau live feed (seperti webcam)

```python
cap = cv2.VideoCapture("example.mp4")

# ambil video dari webcam
# cap = cv2.VideoCapture(0)
```

```python
# tampilkan video

while cap.isOpened():
    ret, frame = cap.read()
    
    # Jika Frame masih tersedia, maka ret = True 
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    cv2.imshow('frame', frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# ambil frames video

cap = cv2.VideoCapture("example.mp4") 

try:  
    # creating a folder named frame-colection 
    if not os.path.exists('frame-collection'): 
        os.makedirs('frame-collection')
except OSError: 
    print ('Error: Creating directory of frame-collection')

frameCounter = 0
while cap.isOpened():
    ret, frame = cap.read()
    
    # Jika Frame masih tersedia, maka ret = True 
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    
    name = './frame-collection/frame' + str(frameCounter) + '.jpg'
    print ('Creating...' + name) 

    # writing the extracted images 
    cv2.imwrite(name, frame) 
    
    frameCounter += 1
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

Untuk menyimpan video diperlukan objek VideoWriter dan nama dari file yang akan di save. Lalu user perlu menentukan FourCC code, nilai FPS dan frame size, dan isColor Flag (jika true maka encoder akan memproses warna)

FourCC adalah kode sepanjang 4- byte yang digunakan untuk mementikan codec video. Berikut adalah daftar code yang bisa ditemukan di fourcc.org. FourCC code tergantung dengan platform yang digunakan.

*    Fedora: DIVX, XVID, MJPG, X264, WMV1, WMV2. (XVID is more preferable. MJPG results in high size video. X264 gives very small size video)
*    Windows: DIVX (More to be tested and added)
*    OSX: MJPG (.mp4), DIVX (.avi), X264 (.mkv).

FourCC code ditulis seperti berikut <code>cv.VideoWriter_fourcc('M','J','P','G')</code> atau <code>cv.VideoWriter_fourcc(*'MJPG')</code>

```python
# SAVE VIDEO FILE

cap = cv2.VideoCapture("example.mp4")

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out =  cv2.VideoWriter('output_flipped.avi', fourcc, 30.0, (1280, 720))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    
    frame = cv2.flip(frame, 0)
#     frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    
    # untuk memberikan text di dalam video
    cv2.putText(frame,"Hello World!!!", (100,100), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255))
    frame = cv2.rectangle(frame, (100,100), (200,200), (0,255,0), 2)

    cv2.imshow('frame', frame)
    
    
    # write the flipped frame
    out.write(frame)
    
    if cv2.waitKey(1) == ord('q'):
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```
